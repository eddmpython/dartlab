## scan() -- 3-step pattern

- `scan()` -- guide (all axes)
- `scan("axis")` -- item list for that axis
- `scan("axis", target)` -- execute

```python
# step 1: guide -- all axes + usage
dartlab.scan()
```

```python
# step 2: item list
dartlab.scan("ratio")
```

```python
# step 3: execute -- rank all companies by ROE
dartlab.scan("ratio", "roe")
```
