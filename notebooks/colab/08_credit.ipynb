```python
import dartlab

c = dartlab.Company("005930")  # Samsung Electronics
```

## 1. Basic credit grade

```python
cr = c.credit()
print(f"Grade: {cr['grade']}")
print(f"Health Score: {cr['healthScore']}/100")
print(f"PD Estimate: {cr['pdEstimate']}%")
print(f"Outlook: {cr['outlook']}")
```

## 2. Detail mode -- 7-axis narratives

```python
cr = c.credit(detail=True)
for n in cr["narratives"]["axes"]:
    print(f"[{n['severity']}] {n['axis']}: {n['summary']}")
```

## 3. Metrics history

```python
cr = c.credit(detail=True)
for h in cr["metricsHistory"][:3]:
    print(
        f"{h['period']}: "
        f"ICR={h.get('ebitdaInterestCoverage', '-')} "
        f"D/E={h.get('debtRatio', '-')}% "
        f"OCF/Sales={h.get('ocfToSales', '-')}%"
    )
```

## 4. Multi-company comparison

```python
for code, name in [("000660", "SK Hynix"), ("035420", "NAVER"), ("003550", "LG")]:
    cc = dartlab.Company(code)
    r = cc.credit()
    if r:
        print(f"{name}: {r['grade']}  Health {r['healthScore']}/100  PD {r['pdEstimate']}%")
```
