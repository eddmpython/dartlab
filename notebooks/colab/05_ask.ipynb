## Provider Setup

`ask()` requires an LLM provider. Choose one:

| Provider | Auth | Cost | Setup |
|----------|------|------|-------|
| `oauth-codex` | ChatGPT subscription (Plus/Team/Enterprise) | Included | `dartlab setup oauth-codex` |
| `gemini` | API key | Free tier | `dartlab setup gemini` |
| `groq` | API key | Free tier | `dartlab setup groq` |
| `openai` | API key (`OPENAI_API_KEY`) | Pay-per-token | `dartlab setup openai` |
| `ollama` | Local install | Free | `dartlab setup ollama` |

`oauth-codex` is recommended -- if you have a ChatGPT subscription, it works with no API keys.

Run `dartlab setup` in terminal to configure, or set inline:

```python
import dartlab

# option 1: use configured default (dartlab setup)
# option 2: override inline
# dartlab.ask("...", provider="gemini")
```

## Basic -- ask a question

```python
# streams to stdout, returns full text
answer = dartlab.ask("삼성전자 재무건전성 분석해줘")
```

## Provider override

```python
# specify provider + model
answer = dartlab.ask("삼성전자 수익성 추세", provider="gemini")
```

## Data filtering

```python
# include only specific data modules
answer = dartlab.ask("삼성전자 핵심 포인트", include=["BS", "IS"])
```

## Analysis pattern

```python
# framework-guided analysis
answer = dartlab.ask("삼성전자 분석", pattern="financial")
```

## Agent mode (chat)

```python
# agent mode -- LLM generates dartlab code for deeper analysis
answer = dartlab.chat("005930", "배당 추세를 분석하고 이상 징후를 찾아줘")
```

## Company bound

```python
c = dartlab.Company("005930")
answer = c.ask("영업이익률 추세 분석해줘")
```
