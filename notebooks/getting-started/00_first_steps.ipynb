## 삼성전자 불러오기

종목코드 6자리를 넣으면 그 회사의 공시 데이터를 자동으로 다운로드한다. 첫 실행은 10~30초 걸린다.

```python
import dartlab

c = dartlab.Company("005930")  # 삼성전자
c.corpName
```

## 재무상태표 (BS)

회사가 가진 자산, 빚, 자본을 보여준다.

```python
c.BS
```

## 손익계산서 (IS)

매출, 영업이익, 순이익 — 얼마 벌고 얼마 남겼는지 보여준다.

```python
c.IS
```

## 현금흐름표 (CF)

실제로 현금이 어디서 들어오고 어디로 나갔는지 보여준다.

```python
c.CF
```

## 재무비율

수익성, 안정성, 성장성 등 50개 비율을 자동 계산한다.

```python
c.ratios
```

## 다른 회사 해보기

종목코드만 바꾸면 된다. 네이버 금융에서 종목을 검색하면 URL에 6자리 코드가 나온다.

```python
sk = dartlab.Company("000660")  # SK하이닉스
sk.IS
```
