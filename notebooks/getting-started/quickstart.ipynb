DartLab은 공시 문서에서 하나의 회사 맵을 만드는 Python 라이브러리다. 한국 DART와 미국 EDGAR를 모두 지원한다.

핵심은 `sections`다. 회사의 모든 공시 섹션을 topic × period 매트릭스로 수평화한다.

## Company 생성

종목코드 또는 회사명으로 생성한다. DART(한국)와 EDGAR(미국) 모두 같은 인터페이스다.

```python
import dartlab

c = dartlab.Company("005930")  # 삼성전자
c.name
```

```python
# 회사명으로도 가능
kakao = dartlab.Company("카카오")
kakao.name
```

## sections — 회사 전체 맵

`sections`는 Polars DataFrame이다. 모든 공시 섹션이 topic × period로 들어 있다.
- `blockType`: text / table 구분
- `textNodeType`: heading / body 구분 (텍스트 구조)
- 기간 컬럼: 최신 먼저, 연간은 Q4 표시

```python
# 회사의 전체 공시 맵
c.sections
```

```python
# topic 목록
c.topics
```

## show(topic) — topic 열기

`show()`는 source 우선순위를 적용한다: finance > report > docs.

```python
# 재무상태표 (finance source)
c.show("BS")
```

```python
# 서술형 topic → 블록 목차
c.show("companyOverview")
```

```python
# 블록 번호 → 실제 데이터
c.show("companyOverview", 0)
```

## trace(topic) — 출처 확인

어떤 source(docs / finance / report)가 채택됐는지 확인한다.

```python
c.trace("BS")
```

```python
c.trace("companyOverview")
```

## diff() — 텍스트 변화 감지

기간 간 서술형 텍스트 변화를 3가지 모드로 감지한다.

```python
# 전체 topic별 변경률
c.diff()
```

```python
# 특정 topic 기간별 이력
c.diff("businessOverview")
```

## 재무제표

재무제표도 sections 위의 finance 뷰다.

```python
c.BS  # 재무상태표
```

```python
c.IS  # 손익계산서
```

```python
c.CF  # 현금흐름표
```

## 재무비율

47개 비율을 6개 카테고리(수익성, 안정성, 성장성, 효율성, 현금흐름, 밸류에이션)로 제공한다.

```python
c.ratios
```

```python
# 비율 시계열
c.finance.ratioSeries
```

## 인사이트

7개 영역(성과, 수익성, 건전성, 현금흐름, 지배구조, 리스크, 기회)의 등급 분석.

```python
ins = c.insights
if ins is not None:
    print(ins.grades())
```

## source namespace

더 깊게 내려가야 할 때는 source namespace를 직접 쓴다.

```python
c.docs.sections       # pure docs source
c.finance.BS          # finance 엔진 직접
c.report.extract("배당")  # report 엔진 직접
```

## EDGAR (미국)

동일한 인터페이스, 다른 데이터 소스. 티커를 넣으면 EDGAR 엔진이 자동 선택된다.

```python
us = dartlab.Company("AAPL")
us.sections
```

```python
us.show("10-K::item1Business")
```

```python
us.BS
```

```python
us.ratios
```
