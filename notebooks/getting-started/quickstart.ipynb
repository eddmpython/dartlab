## Company 생성종목코드 또는 회사명으로 생성한다. DART(한국)와 EDGAR(미국) 모두 같은 인터페이스다.

```python
import dartlabc = dartlab.Company("005930")  # 삼성전자c.name
```

## sections — 회사 전체 맵`sections`는 Polars DataFrame이다. 모든 공시 섹션이 topic × period로 들어 있다.- `blockType`: text / table 구분- `textNodeType`: heading / body 구분 (텍스트 구조)- 기간 컬럼: 최신 먼저, 연간은 Q4 표시

```python
# 회사의 전체 공시 맵c.sections
```

```python
# topic 목록c.topics
```

## show(topic) — topic 열기`show()`는 source 우선순위를 적용한다: finance > report > docs.

```python
# 재무상태표 (finance source)c.show("BS")
```

```python
# 서술형 topic → 블록 목차c.show("companyOverview")
```

```python
# 블록 번호 → 실제 데이터c.show("companyOverview", 0)
```

## trace(topic) — 출처 확인어떤 source(docs / finance / report)가 채택됐는지 확인한다.

```python
c.trace("BS")
```

```python
c.trace("companyOverview")
```

## diff() — 텍스트 변화 감지기간 간 서술형 텍스트 변화를 3가지 모드로 감지한다.

```python
# 전체 topic별 변경률c.diff()
```

```python
# 특정 topic 기간별 이력c.diff("businessOverview")
```
