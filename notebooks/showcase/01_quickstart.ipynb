**DartLab** turns Korean (DART) and US (SEC EDGAR) filings into a unified company map.

In this notebook you'll learn:
- Create a Company in one line (Korean or US)
- `sections` — the full company map (topic × period)
- `show(topic)` — open any topic
- `diff()` — detect what changed between periods
- Financial shortcuts: `BS`, `IS`, `CF`, `ratios`

## 1. Create a Company

DartLab auto-detects whether it's a Korean stock code, company name, or US ticker.

```python
import dartlab

# Korean company — by stock code
samsung = dartlab.Company("005930")
print(f"Company: {samsung.corpName} ({samsung.stockCode})")
```

```python
# US company — by ticker
apple = dartlab.Company("AAPL")
print(f"Company: {apple.corpName} ({apple.ticker})")
```

## 2. Sections — The Full Company Map

`sections` horizontalizes every filing section into a **topic × period** matrix.
This is dartlab's core innovation — no other tool does this.

```python
# Every topic, every period, in one DataFrame
samsung.sections
```

```python
# What topics are available?
samsung.topics
```

## 3. Show — Open Any Topic

`show(topic)` applies source priority: **finance > report > docs**.

```python
# Financial statement (finance source, standardized)
samsung.show("BS")
```

```python
# Narrative topic (docs source)
samsung.show("businessOverview")
```

## 4. Diff — Detect What Changed

dartlab compares filing text across periods — **something Bloomberg can't do**.

```python
# Overview: which topics changed the most?
samsung.diff()
```

```python
# Drill into a specific topic
samsung.diff("riskManagement")
```

## 5. Financial Shortcuts

Standardized financial statements — same `snakeId` across all companies.

```python
samsung.BS  # Balance Sheet
```

```python
samsung.IS  # Income Statement
```

```python
samsung.CF  # Cash Flow
```

```python
# 47 financial ratios (profitability, stability, growth, efficiency, cashflow, valuation)
samsung.ratios
```

## 6. Search & Listing

```python
# Search Korean companies
dartlab.search("반도체")
```

```python
# Full listing (2,700+ Korean companies)
dartlab.listing()
```

## 7. Same Interface for US Companies

```python
apple = dartlab.Company("AAPL")

# Same API, different market
apple.sections
```

```python
apple.BS
```

```python
apple.IS
```
