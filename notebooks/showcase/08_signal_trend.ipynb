## Why This Matters

This is **alternative data from public filings**:

- When companies start mentioning "AI" more in filings, it signals industry-wide adoption
- A spike in "supply chain" mentions = market-wide risk awareness
- "M&A" keyword frequency correlates with deal activity

Unlike news sentiment, **filing text is legally binding** — companies can't exaggerate or omit.
This makes disclosure signals more reliable than news-based sentiment.
