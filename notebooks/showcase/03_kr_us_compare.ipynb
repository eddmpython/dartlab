DartLab is the **only library** that lets you compare Korean (DART) and US (EDGAR) companies with the same interface.

In this notebook:
- Samsung vs Apple — financial statements side by side
- Standardized ratios comparison
- Disclosure sections comparison (risk factors, business overview)
- SK Hynix vs Micron — semiconductor peers

````python
!pip install -q dartlab
````

````python
import dartlab

samsung = dartlab.Company("005930")  # Samsung Electronics (Korea)
apple = dartlab.Company("AAPL")      # Apple Inc. (US)
````

## 1. Income Statement — Side by Side

Both use the same `snakeId` columns (`revenue`, `operating_income`, `net_income`).

````python
print("=== Samsung Electronics ===")
samsung.IS
````

````python
print("=== Apple Inc. ===")
apple.IS
````

## 2. Balance Sheet Comparison

````python
print("=== Samsung BS ===")
samsung.BS
````

````python
print("=== Apple BS ===")
apple.BS
````

## 3. Financial Ratios

47 standardized ratios. Compare profitability, stability, growth at a glance.

````python
print("=== Samsung Ratios ===")
samsung.ratios
````

````python
print("=== Apple Ratios ===")
apple.ratios
````

## 4. Sections — Filing Structure Comparison

Both DART annual reports and SEC 10-K filings are horizontalized into `sections`.

````python
# Samsung: DART annual report sections
samsung.topics
````

````python
# Apple: SEC 10-K sections
apple.topics
````

## 5. Semiconductor Peers: SK Hynix vs Micron

````python
skhynix = dartlab.Company("000660")  # SK Hynix
micron = dartlab.Company("MU")        # Micron Technology

print(f"SK Hynix: {skhynix.corpName}")
print(f"Micron: {micron.corpName}")
````

````python
# Compare income statements — memory chip rivals
print("=== SK Hynix IS ===")
skhynix.IS
````

````python
print("=== Micron IS ===")
micron.IS
````

## Key Insight

DartLab unifies **DART (Korea)** and **EDGAR (US)** into one `Company` interface:

```python
dartlab.Company("005930")  # Korea → DART engine
dartlab.Company("AAPL")    # US → EDGAR engine
```

Same methods, same column names, same data structure — just different markets.
No other open-source tool can do this.
