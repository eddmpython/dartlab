The **sections** architecture is DartLab's core innovation.
It horizontalizes every filing into a `topic × period` matrix — the full company map.

In this notebook:
- `sections` — the full company map
- 4 namespaces: docs, finance, report, profile
- Source tracking with `trace()`
- Disclosure-specific topics (risk, MD&A, segments)
- Filing list with `filings()`

````python
import dartlab

c = dartlab.Company("005930")  # Samsung Electronics
````

## 1. Sections — The Full Company Map

`c.sections` = `c.profile.sections` (merged canonical view).
This is where finance replaces numerical topics and report fills in structured data.

````python
c.sections
````

````python
# Pure docs view (before finance/report replacement)
c.docs.sections
````

## 2. Topics — What's Available?

````python
c.topics
````

## 3. Show — Open Any Topic

Source priority: **finance > report > docs**.

````python
# Employee overview
c.show("employeeOverview")
````

````python
# Dividend information
c.show("dividend")
````

## 4. Trace — Where Did This Data Come From?

````python
# Financial data comes from the finance engine (XBRL)
c.trace("BS")
````

````python
# Narrative data comes from docs (filing text)
c.trace("businessOverview")
````

## 5. Filings — Document List

````python
c.filings()
````

## 6. The 4 Namespaces

| Namespace | Role | Authoritative for |
|-----------|------|-------------------|
| `c.docs` | Pure filing text | Narrative, risk, MD&A |
| `c.finance` | XBRL numbers | BS, IS, CF, ratios |
| `c.report` | Structured API data | 28 report categories (DART only) |
| `c.profile` | Merged view | Everything (default) |

````python
# docs namespace — pure filing source
c.docs.sections
````

````python
# finance namespace — authoritative numbers
c.finance.IS
````

## Architecture Summary

```
docs.sections (topic × period — pure text)
  ↓ finance replaces numerical topics
  ↓ report fills structured data (DART only)
profile.sections (merged canonical view) = c.sections
```

This is what makes DartLab unique: **every filing, every period, one unified map.**
No other tool horizontalizes narrative + financial + structured data this way.
