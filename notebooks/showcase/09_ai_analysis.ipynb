## 4. How It Works — 2-Tier Architecture

```
Tier 1 (System-driven, all providers):
  Question → Intent classification → Analysis package selection
  → Parallel engine computation → Context assembly → LLM response

Tier 2 (LLM-driven, tool-calling providers only):
  LLM judges "not enough" → Calls low-level tools
  → Raw timeseries, disclosure text search, multi-company comparison
```

**Key principle**: LLM only interprets and explains. DartLab does all calculation.
This prevents hallucination on financial numbers.
