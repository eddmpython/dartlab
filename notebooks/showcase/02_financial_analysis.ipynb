DartLab normalizes XBRL financial data into a **standardized `snakeId` schema** — the same IDs across all 2,700+ Korean companies and US companies.

In this notebook:
- Standardized financial statements (BS, IS, CF)
- Time series for any metric
- 47 financial ratios (profitability, stability, growth, efficiency, cashflow, valuation)
- Sector benchmark comparison

```python
import dartlab

c = dartlab.Company("005930")  # Samsung Electronics
c.corpName
```

## 1. Financial Statements

Each statement returns a **quarterly standalone** time series DataFrame.
Account IDs are normalized to `snakeId` (e.g., `revenue`, `operating_income`, `total_assets`).

```python
# Balance Sheet — point-in-time balances
c.BS
```

```python
# Income Statement — quarterly standalone (cumulative → standalone conversion)
c.IS
```

```python
# Cash Flow Statement — quarterly standalone (delta from prior quarter)
c.CF
```

## 2. Time Series

`timeseries(metric)` returns annual values for any standardized metric.

```python
# Revenue over time
c.timeseries("revenue")
```

```python
# Operating income
c.timeseries("operating_income")
```

```python
# Total assets
c.timeseries("total_assets")
```

## 3. Financial Ratios

47 ratios organized by category: profitability, stability, growth, efficiency, cashflow, valuation.

```python
c.ratios
```

```python
# Ratio time series (multi-year)
c.ratioSeries
```

## 4. Finance Namespace — Full Access

`c.finance` is the authoritative source for numerical data.

```python
# The finance namespace
c.finance.BS   # same as c.BS
c.finance.IS   # same as c.IS
c.finance.CF   # same as c.CF
```

## 5. Sector Benchmark

Compare any company's ratios against its sector (P10 / median / P90).

## Key Insight

DartLab's `snakeId` normalization maps **34,175 different XBRL account names** to ~100 standard IDs.
This means you can compare Samsung's `revenue` with Apple's `revenue` — same column name, same semantics.
