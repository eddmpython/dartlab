detecting additions, deletions, and changes at the line and character level.

In this notebook:
- Detect which topics changed the most
- Drill into risk factor changes
- Line-by-line diff with change classification
- Track narrative shifts across years

```python
!pip install -q dartlab
```

```python
import dartlab

c = dartlab.Company("005930")  # Samsung Electronics
```

## 1. Overview — Which Topics Changed?

`diff()` with no arguments returns a summary: change rate, added/removed counts per topic.

```python
# Which sections had the most changes?
c.diff()
```

## 2. Topic History — Period-by-Period Changes

Pass a topic name to see how it changed across all periods.

```python
# Risk management changes across periods
c.diff("riskManagement")
```

```python
# Business overview changes
c.diff("businessOverview")
```

## 3. Detailed Diff — Line by Line

Specify two periods to get the actual text diff.

```python
# Line-level diff between two annual reports
c.diff("riskManagement", "2023", "2024")
```

## 4. Source Tracking

`trace()` shows which data source was used for each topic.

```python
c.trace("riskManagement")
```

## 5. US Company Diff — 10-K Risk Factors

```python
apple = dartlab.Company("AAPL")
apple.diff()
```

## Why This Matters

Analysts manually compare filings year-over-year. DartLab automates this:

- **3-stage diff**: hash detection → line diff → character diff
- **7 change types**: unchanged, edited, restated, moved, added, removed, placeholder
- **Vectorized**: Polars xxhash64 — handles 100K+ cells in milliseconds

Use cases:
- "Did Samsung add new AI regulation risk this year?"
- "What changed in the supply chain risk section?"
- "How did the MD&A narrative shift after the earnings miss?"
