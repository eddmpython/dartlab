## 2. Group Health Analysis

Crosses the network graph with financial ratios to find weak links in conglomerate groups.

```python
summary, weak_links = dartlab.groupHealth()
print("=== Group Summary ===")
summary
```

## Why This Matters

Korean conglomerates (chaebols) have complex cross-ownership structures.
DartLab's network engine automatically extracts these relationships from filings
and highlights **financial weak links** — groups where one troubled subsidiary
could cascade risk to the parent.
