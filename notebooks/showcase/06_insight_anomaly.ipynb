## 1. Insights — 7-Area Grading

| Area | What it measures |
|------|------------------|
| performance | Revenue/OP YoY + quarterly volatility |
| profitability | Operating margin, net margin, ROE, ROA |
| health | Debt ratio, current ratio, interest coverage |
| cashflow | Operating CF, FCF, cash position |
| governance | Largest shareholder, audit opinion, outside directors |
| risk | Cross-area risk flags |
| opportunity | Cross-area opportunity flags |

```python
c.insights
```

## 2. Anomaly Detection

6 automated rules that flag suspicious financial patterns:

| Rule | What it detects |
|------|----------------|
| earnings_quality | Operating income ↑ but operating CF ↓ |
| working_capital | Receivables/inventory growing faster than revenue |
| balance_sheet_shift | Debt/equity structure changed ≥50% |
| cash_burn | Cash declining + negative operating CF |
| margin_divergence | Operating margin shifted ≥5%p |
| financial_sector | Banking-specific anomalies |
