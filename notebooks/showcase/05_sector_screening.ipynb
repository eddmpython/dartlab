## 1. Stock Screening — 8 Presets

| Preset | English | Criteria |
|--------|---------|----------|
| 가치주 | Value | ROE≥10, Debt≤100, PBR≤1.5 |
| 성장주 | Growth | Revenue growth≥15%, OP growth≥20% |
| 턴어라운드 | Turnaround | Prior loss → current profit |
| 현금부자 | Cash-rich | FCF positive, Cash/Assets≥20% |
| 고위험 | High-risk | Debt≥200, ICR<3 |
| 자본잠식 | Capital erosion | Equity shrinking |
| 소형고수익 | Small-cap high-profit | Small cap + high margins |
| 대형안정 | Large-cap stable | Top 100 by assets, stable margins |

```python
# Value stocks
dartlab.screen("가치주")
```

```python
# Growth stocks
dartlab.screen("성장주")
```

```python
# Cash-rich companies
dartlab.screen("현금부자")
```

```python
# High-risk (caution)
dartlab.screen("고위험")
```
