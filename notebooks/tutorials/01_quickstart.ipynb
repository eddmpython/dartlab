DartLab의 핵심 흐름: **sections → show → trace → diff**

이 노트북에서 배우는 것:
- `c.sections` — 회사 전체 맵
- `c.show(topic)` — topic 하나 열기
- `c.trace(topic)` — 출처 확인
- `c.diff()` — 텍스트 변화 감지
- `c.BS / IS / CF / ratios` — 재무 바로가기

## Company 생성

```python
import dartlab

c = dartlab.Company("005930")  # 삼성전자
c.name
```

```python
# 회사명으로도 가능
kakao = dartlab.Company("카카오")
kakao.name
```

## sections — 회사 전체 맵

`c.sections`는 회사의 모든 공시 섹션을 topic × period로 수평화한 DataFrame이다.

```python
c.sections
```

```python
# topic 목록 확인
c.topics[:10]  # 처음 10개
```

## show — topic 열기

`show(topic)`은 source 우선순위(finance > report > docs)를 적용한다.

```python
# 재무상태표 (finance source)
c.show("BS")
```

```python
# 서술형 topic → 블록 목차
c.show("companyOverview")
```

```python
# 블록 번호 → 실제 데이터
c.show("companyOverview", 0)
```

## trace — 출처 확인

```python
c.trace("BS")
```

## diff — 텍스트 변화 감지

```python
# 전체 변경률
c.diff()
```

```python
# topic별 이력
c.diff("businessOverview")
```
