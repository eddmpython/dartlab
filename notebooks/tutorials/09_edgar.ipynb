## 1. EDGAR Company 생성

미국 종목은 티커로 생성한다. DartLab이 자동으로 EDGAR 엔진을 선택한다.

```python
import dartlab

c = dartlab.Company("AAPL")
c.corpName, c.ticker, c.market
```

## 2. index — 사용 가능한 데이터 확인

DART Company와 동일한 8컬럼 구조를 반환한다.

```python
c.index
```

## 3. 재무제표 조회

DART Company와 동일한 shortcut property로 접근한다.

```python
c.BS
```

```python
c.IS
```

```python
c.CF
```

## 4. 10-K 서술형 섹션

`show()`는 공시 topic에 대해 `ShowResult(text, table)`을 반환한다.

```python
result = c.show("riskFactors")
result.text
```

## 5. docs.sections — 10-K/10-Q 수평화

모든 10-K/10-Q 섹션을 기간별로 수평 정렬한 DataFrame이다.

```python
c.docs.sections
```

## 6. source 추적

```python
c.trace("BS")
```

```python
c.trace("riskFactors")
```

## 7. DART Company와 비교

같은 `index → show → trace` 흐름이 한국 종목에도 동일하게 적용된다.

```python
kr = dartlab.Company("005930")
kr.corpName, kr.stockCode
```

```python
kr.index
```

```python
kr.show("BS")
```
